### Data Understanding
The data has 16 columns and 20058 observations.

The variables in the dataset are as follows:

1. id – Unique Identification Number of a game.

2. rated – Whether a game is rated or casual.

3. created_at – The time recorded when the game has started.

4. last_move_at – The time recorded when the game has finished.

5. turns – Number of turns in a game.

6. victory_status – In what way the victory or an outcome of a game is obtained and there are four types of victory_status viz. ‘check mate’, ‘out of time’, ‘resign’, and ‘draw’.

7. winner – Outcome of a game i.e., black piece player’s win, white piece player’s win and draw.

8. Increment_code – Increment code is time expressed in the form ‘5+8’ which means the total time for each player is 5 minutes and for every move by a player, extra 8 seconds are added to the player's total time.

9. white_id – ID of white player.

10. white_rating – Rating of white player.

11. black_id – ID of black player.

12. black_rating – Rating of black player.

13. moves – Every move in a game by both players (Scorecard of a game).

14. opening_eco – Encyclopedia of Chess Openings which describes and classifies the openings of a chess game.

15. opening_name – The name of the opening in a chess game.

16. opening_ply – The number of turns played by both players in an opening game.

I am going to predict the probability of winning on the basis of first move of the both black and white player.

```python
#Identify the first moves of Both white and black 
moves = df.moves.str.split(n=2,expand=True)
moves
```

Splited the moves into three column, first column contains the first move of white player and second column contains the first move of black player and third column contains the rest of the moves which I am not going to consider into my model because it will take more time to do data manipulation I will do it later

```python
#creating column for first move of the white
df['white_1move'] = moves[0]
#creating column for first move of the black
df['black_1move'] = moves[1]
```

```python
df.head()
```

```python
#Finding probability of winning white
df_wm = pd.crosstab(df.white_1move,columns = df.winner,normalize='index')
df_wm.head()
```

```python
dict(df_wm['white'])
```

```python
#Mapping white first moves with the probability of winning white
df['white_1move'] = df['white_1move'].map(dict(df_wm['white']))
```

```python
df.head()
```

```python
#probability of winning of white player on the basis of first move of black player
df_bm = pd.crosstab(df.black_1move,columns = df.winner,normalize='index')
df_bm.head()
```

```python
#Mapping black first moves with the probability of winning black W.R.T white
df['black_1move'] = df['black_1move'].map(dict(df_bm['white']))
```

```python
df.head()
```

```python
df_ = pd.crosstab(df.opening_eco,columns = df.winner,normalize='index')
df_.head()
```

```python
df['opening_eco'] = df['opening_eco'].map(dict(df_['white']))
```

```python
df.head()
```

```python
#Dropping moves and victory_status columns
df.drop(['moves','victory_status'],axis=1,inplace=True)
```

```python
df.dtypes
```

### Splitting the dataset

```python
#Identify Input and Output columns
X = df.drop('winner',axis=1)
y = df.winner
```

```python
#check the shape of input data
X.shape
```

```python
#check the shape of output data
y.shape
```

```python
#Split the data into Train and Test sets
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
#check the shape of train and test sets
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit(X_train)
```

```python
X_train_scaled = sc.transform(X_train)
X_test_scaled = sc.transform(X_test)
```

### Building Models:
1. Logistic Regression
2. SVM ==> SVC
3. Random Forest

#### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression()
```

```python
model1.fit(X_train_scaled,y_train)
```

```python
y_pred1 = model1.predict(X_test_scaled)
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param = [
    {'n_estimators' : [3,10,100],
    'criterion' : ['gini','entropy'],
    'min_samples_split' : [2,4,6],
    'min_samples_leaf' : [1,2,3],
    'max_features' :[4,8,12]}
]
```
